# DATA CLEANING

```python
#Required packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
# Createing data frame
pd.set_option("display.max_rows", 50)
# Suppress warnings
warnings.filterwarnings('ignore')
```

```python
loan = pd.read_csv("./loan/loan.csv")
```

```python
# Print head
loan.head()
```

```python
loan.shape
```

```python
# get details of the data
loan.info()
```

```python
# Find null values in dataset and get the percentage of coloumns having missing columns
missing_values_percentage = loan.isnull().sum() / len(loan) * 100
print(missing_values_percentage)
```

```python
# Get coloumns with more than 40% of data
columns_with_missing_data = missing_values_percentage[missing_values_percentage > 40].index
columns_with_missing_data
```

```python
# Drop the columns with more than 40% of miss_dataing data
loan_data = loan.drop(columns=columns_with_missing_data)
```

```python
# Get the shape and find the number of columns after dropping columns 
loan_data.shape
```

```python
# Checking for the coloumns that still have some missing values
missing_value = loan_data.isnull().sum() / len(loan_data) * 100
find_missing_columns = missing_value[missing_value > 0]
print(find_missing_columns)
```

```python
# As we have reduce to 54 columns after data cleaning. Now we will again check for the columns with the unique counts in them.
unique_col = loan_data.nunique()
print(unique_col)
```

```python
# As per the above observation we got the columns with single unique value, so they will not affect much in our analysis
#remove the columns with unique count = 1
uni_col = unique_col[unique_col.values == 1]
unique_col_list = uni_col.index.tolist()
loan_data.drop(columns=unique_col_list, inplace=True)
```

```python
#Checking the number of the columns left after removing columns having uniques equal to 1.
loan_data.shape
```

```python
# After observing the data found that few columns will not  help in the analysis with following reasons so droping them
cols_not_required = ["url", "desc", "out_prncp", "out_prncp_inv", "id", "member_id"]
loan_data.drop(columns=cols_not_required, inplace=True)
```

```python
# Now again check for the columns let with missinn data
col_with_missing_data = round(100*(loan_data.isnull().sum()/len(loan_data)), 2)
col_with_missing_data[col_with_missing_data != 0]
```

```python
#get the unique numbers of values in columns with missing data
for col in col_with_missing_data[col_with_missing_data != 0].index.tolist():
    print(f"Numbers of unique counts in {col}: {len(loan_data[col].unique())}")
```

```python
# get the unique value count in each above columns
for col in col_with_missing_data[col_with_missing_data != 0].index.tolist():
    print(f"Unique value count in each {col}: {loan_data[col].value_counts()}")
```

```python
# As per the above analysis pub_rec_bankruptcies seem to have 0 as more value counts
loan_data = loan_data[~loan_data.pub_rec_bankruptcies.isnull()]
```

```python
#Remove rows where the loan_status is current as current means in-porgress
rows_to_drop = loan_data[loan_data['loan_status'] == "Current"].index
loan_data.drop(rows_to_drop, inplace=True)
```

```python
#Check for columns still left with missin counts
missing_val_col = round(100*(loan_data.isnull().mean()), 2)
missing_val_col[missing_val_col > 0]
missing_val_col
```

```python
# Found "emp_length", "last_pymnt_d", "revol_util", "emp_title", "title" left with some missing values
# Let's work in fixing the missing values for these columns

# emp_length and emp_title have around 6.19 and 2.73% of missing values, these two columns contains brower information of years and job title can't be imputed directly
#so removing the rows with null values

loan_data = loan_data[~loan_data["emp_length"].isnull()]
loan_data = loan_data[~loan_data["emp_title"].isnull()]
loan_data = loan_data[~loan_data["last_pymnt_d"].isnull()]
loan_data = loan_data[~loan_data["revol_util"].isnull()]
loan_data = loan_data[~loan_data["title"].isnull()]

```

```python
#Checking shape again 
loan_data.shape
```

```python
#Checking still the data is missing or not
loan_data.isnull().sum()
```

```python
loan_data.to_csv("cleand_data.csv")
```

# Data Preparation

```python
duplicate_rows = loan_data.duplicated(keep=False)
duplicate_rows
```

```python
loan_data.info()
```

```python
#By looking at the data, divided the columns into their category type
date_columns = ["issue_d", "earliest_cr_line", "last_pymnt_d", "last_credit_pull_d"]

categorical = ["grade", "sub_grade", "emp_title", "home_ownership", "verification_status", "loan_status", "purpose", "title", "zip_code", "addr_state"]
numerical = ["loan_amnt", "funded_amnt", "funded_amnt_inv", "term", "int_rate", "installment", "emp_length", "annual_inc", "dti", "delinq_2yrs", "inq_last_6mths", "open_acc", "pub_rec", "revol_bal", "revol_util", "total_acc", "total_pymnt", "total_pymnt_inv", "total_rec_prncp", "total_rec_int", "total_rec_late_fee", "recoveries", "collection_recovery_fee", "last_pymnt_d", "last_pymnt_amnt", "last_credit_pull_d",  "pub_rec_bankruptcies"]
```

```python
#Now check for the dtype of the attributes
loan_data[date_columns].info()
```

```python
loan_data[categorical].info()
```

```python
loan_data[numerical].info()
```

```python
for col in date_columns:
    loan_data[col] = pd.to_datetime(loan_data[col], format = '%b-%y')
```

```python
loan_data[date_columns].info()
```

```python
#changing percentage
per_col = ["int_rate", "revol_util"]
for col in per_col:
    loan_data[col] = loan_data[col].str.strip('%').astype('float')
```

```python
loan_data[per_col].info()
```

```python
#Manipulating employee length
loan_data.emp_length.value_counts()
```

```python
emp_len_conversion = { '< 1 year' : 0, '1 year' : 1, '2 years' : 2, '3 years' : 3, '4 years' : 4, '5 years' : 5, '6 years' : 6, '7 years' : 7, '8 years' : 8, '9 years' : 9, '10+ years' : 10 }
loan_data = loan_data.replace({"emp_length": emp_len_conversion })
    
```

```python
loan_data["emp_length"].value_counts()
```

```python
#Manipulating term data
loan_data["term"] = loan_data["term"].apply(lambda x : x.split()[0])
loan_data["term"].value_counts()
```

```python
#Manipulating issue date 
loan_data["issue_month"] = loan_data["issue_d"].dt.month
loan_data["issue_year"] = loan_data["issue_d"].dt.year
```

```python
loan_data["issue_month"].value_counts()
```

```python
loan_data["issue_year"].value_counts()
```

```python
loan_data.to_csv("cleaned_loan_data.csv", encoding="utf-8", index=False)
```
